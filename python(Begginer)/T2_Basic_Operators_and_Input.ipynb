# Operators : 

Bellow are the 4 basic common operators used in python: 

[1]  (+) -> this is the plus sign which is the addition operator.

[2]  (-) -> this is the minus sign which is the subtraction operator.  

[3]  (/) -> this is the division sign which is the division operator.  

[4]  (*) -> this is the multiplication sign which is the multiplication operator.    


When we use these operators it's important that we use them on certain data types.


For example in math I can add two numbers like ( 3 + 4 ) = 7 , so the computer can actually do that math operation by using that operator to return this value of (7).

Examples: 

```python
num1 = 45
num2 = 15
print ( num1 + num2)   # (45) plus (15) = (60) 
print ( num1 - num2)   # (45) minus (15) = (30)
print ( num1 / num2)   # (45) divide (15) = (3.0)
print ( num1 * num2)   # (45) multiply (15) = (675)
```

Here I used the (num1) and (num2) variables to assign numbers for each one and then I used the different operation signs with the varibale names stated above inside the print statment in order to perform the calculation operation and print it to the council 

# Other useful operators to know as well:  

[1]  (**) -> this is the exponent sign operator.

[2]  (//) -> this is the integer division sign operator (used to get the whole number in the division and removes the reminder).  

[3]  (%) -> this is the modulus sign operator (used to get the reminder from the division and not the whole number).   

Examples: 

```python
num3 = 2
num4 = num1 ** num3   # (45) power (2) 
num5 = num1 // num3   # (45) integer division (2)
num6 = num1 % num3    # (45) modulus (2) 

print(num4)   # prints (2025) from --> (45) power (2) 
print(num5)   # prints (22) from --> (45) integer division (2)
print(num6)   # prints (1) from --> (45) modulus (2)
```

Here, I added a new varibale called (num3) and assigned it the value of 2.

After that, I added 3 more varibales (num4) , (num5) , (num6) and for each one of these variables I inserted an operation using the varibles (num1) and (num3).   

Finally, I added the values stored in (num4) , (num5) , (num6) inside a print statment each to print the output for each of these operations on the concil for the user to view.
